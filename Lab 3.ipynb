# Problem 1
a) Use the numpy function [numpy.random.normal](https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html) to generate a standard Gaussian random variable $X$ (zero mean and unit variance).

- Use histogram to plot the estimated PDF of the random variable $X$
- Compute the true "analytical" PDF of $X$ . Plot the true PDF on the top of the histogram figure.
- Calculate the sample mean and the sample variance.

```python
#type your code here:
'''
In This problem we want to make a sample which have a standard normal distribution of (mu=0, sigma=1)
'''
mu, sigma = 0, 1
sample = np.random.normal(mu,sigma,10000)  # Here we construct the sample which has 10000 elements
mean = np.mean(sample) # Here we calculate the mean of the sample
variance = np.var(sample, ddof=1) #Here we grt the variance of the sample

print('Sample mean: ',mean)
print('Sample variance: ',variance)

count, bins, ignored = plt.hist(sample, 50, density=True) # After that we plot the output of the sample in a histogram
# arguments: sample is the elements, bins is the intervals we draw with
# ignored has no meaning as we ignore it but we assign it so as nit to get an error
# density=true to make the bins width as the probability density (it has a documentation in plt.hist)

plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r') # Here we plot the red line which represents the pdf according to the emperical law
# 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ) 
# The above law is the pdf of normal ditribution, google it

plt.show() # To show only the plot with no info. about the plot
```

C) Repeat (a) for an exponential random variable with $\lambda = 1$ using the numpy function [numpy.random.exponential](https://numpy.org/doc/stable/reference/random/generated/numpy.random.exponential.html)

```python
#type your code here:
'''
The same but here we use the exponential distribution, google ( numpy.random.exponential) to get the pdf
'''
lamda = 1
beta = 1/lamda
sample = np.random.exponential(beta, 100000) 
mean = np.mean(sample)
variance = np.var(sample, ddof=1)

print('Sample mean: ',mean)
print('Sample variance: ',variance)

count, bins, ignored = plt.hist(sample, 50, density=True)

plt.plot(bins, lamda*np.exp(-(bins*lamda)),linewidth=2, color='r')
#lamda*np.exp(-(bins*lamda)) is the pdf of exponential distribution
plt.show()
```

# Problem 2: Practicing the Central Limit Theorem
Use Python to simulate the Central Limit Theorem for the following random i.i.d variables and check that the higher the number of number the variables, the more the mean of subsets behave like a gaussian random variable.

a) Exponential random variable with $\lambda = 0.5$

```python
#type your code here:
'''
In this problem we want to use the CLT to construct a normal distribution, google Central Limit Theorem
'''
lamda = 0.5 # here is the parameter of exponential distribution
n = 10000 # No. of experiments (no. of samples we took from population)
s = 5 # No. of elements in each sample, each sample has 5 elements

sample = np.random.exponential(1/lamda, (n,s)) #here we construct the random samples according to the exponential distribution
# (n,s) both represents the size of array, (n gives 10000 rows) meaning we have 10000 sample
# (s gives 5 columns) meaning we have 5 elements in each sample

mean = np.mean(sample, axis=1) #Here we get the mean of each sample which gives us 10000 means in the (mean array)
plt.hist(mean, bins=50) # Here we represent the distribution of means in the histogram
plt.show() # To show only the plot with no info. about the plot

# WHAT WE CAN GET OUT OF ALL OF THIS?
# if we can notice we are here using the exponential distribution, to see hoe it looks like, go to problem-1 (c,d)
# They represent hoe the exponential distribution looks like
# but with Central Limit Theorem we can represent it as a normal distribution
# This helps us to make statistical inferences based on established statistical procedures of the normal distribution
# All these procedures are accomplished in all other distributions, compare each distribution befor and after the CLT
```
